## Executive Summary

This project analyzes the relationship between product price and customer demand to identify a revenue-maximizing pricing strategy.

Using price elasticity modeling and revenue simulation:
- Demand was found to be **relatively inelastic**
- Revenue increases as price increases within the tested range
- Pricing optimization offers a significant opportunity to improve revenue

The analysis demonstrates how data-driven pricing decisions can directly impact business performance.


## Key Findings

1. **Demand is relatively inelastic**
   - Estimated price elasticity of demand: **−0.264**
   - Customers are not highly sensitive to price changes

2. **Price significantly influences demand**
   - The relationship between price and demand is statistically significant (p < 0.001)
   - Price alone explains a large portion of demand variability (R² ≈ 0.86)

3. **Revenue increases with price**
   - Revenue simulation shows higher prices lead to higher total revenue
   - The revenue-maximizing price lies above the current average price


## Revenue Impact

- Current pricing does not maximize revenue
- Simulated pricing scenarios indicate a potential revenue uplift of **~35%**
- Because demand is inelastic, revenue growth is driven more by **pricing strategy** than by increasing sales volume


## Business Recommendation

Based on the analysis, the following pricing strategy is recommended:

- Gradually increase prices while monitoring demand response
- Focus on pricing optimization rather than volume-based growth
- Use controlled experiments (A/B testing) to validate price changes in production

This approach balances revenue growth with customer retention.


## Assumptions & Limitations

- The analysis assumes **constant price elasticity**, which may not hold at extreme price levels
- The dataset is synthetic and does not account for:
  - Competitor pricing
  - Customer segmentation
  - Promotional effects
- The optimal price was identified within a predefined simulation range

In real-world deployment, these factors should be incorporated before final pricing decisions.
