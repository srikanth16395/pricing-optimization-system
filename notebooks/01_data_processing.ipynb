```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)

n = 1000

price = np.random.uniform(10, 100, n)
quantity_sold = 2000 - (price * np.random.uniform(8, 15)) + np.random.normal(0, 50, n)
quantity_sold = np.maximum(quantity_sold, 0)

df = pd.DataFrame({
    "price": price,
    "quantity_sold": quantity_sold
})

df.head()
```

```python
#save raw data
df.to_csv(r"D:\Fogan\Github_Projects\pricing-optimization-system\data\raw\pricing_data.csv", index=False)
```

```python
# EDA
plt.figure(figsize=(6,4))
plt.scatter(df["price"], df["quantity_sold"], alpha=0.6)
plt.xlabel("Price")
plt.ylabel("Quantity Sold")
plt.title("Price vs Quantity Sold")
plt.savefig(r"D:\Fogan\Github_Projects\pricing-optimization-system\images\price_vs_quantity.png")
plt.show()
```
